```python
import numpy as np
import pandas as pd
from copy import deepcopy
from bertopic import BERTopic
```

```python
df = pd.read_excel("Dec.xlsx")
```

```python
df.tail(3)
```

```python
docs = list(df.loc[:, "Column1.text"].values)
```

```python
docs[100]
```

```python
model = BERTopic(language="english")
```

```python
topics, probs = model.fit_transform(docs)
```

```python
model.get_topic_freq()
```

```python
model.get_topic(1)
```

```python
df = pd.DataFrame({'topic': topics, 'document': docs})
```
